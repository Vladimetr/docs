```python
# -------------------------- Круговая диаграмма вероятностей ---------------------
import matplotlib.pyplot as plt
plt.switch_backend('agg') # !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
import numpy as np

a = [0, 0, 0]
a = [w+10 for w in a]
print(a)

labels = ['rus', 'eng', 'enother']
procents = [35.55, 55.4, 10.78]
#colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
lbl_max = np.argmax(procents)
explode = [0, 0, 0]
for i in range(3):
    if i == lbl_max:
        explode[i] = 0.2
explode = tuple(explode)

fig = plt.figure(figsize=(15, 10))
patches, lbls_text, proc_text = plt.pie(procents, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow=True, startangle=90) # + colors=colors

# настройка шрифтов для надписей
plt.setp(lbls_text, size=15, weight='bold')
plt.setp(proc_text, size=13)

fig.savefig('out.png')

# comment this if you want to Display
#plt.close()
```

```python
# -------------------------------- A few line in graphic -------------------------------

import matplotlib.pyplot as plt
import numpy as np
from math import sin
import matplotlib
matplotlib.rcParams['figure.figsize'] = (15.0, 10.0)  # 15 * 80 pixels along X
                                                      # 10 * 80 pixels along Y

x = np.arange(0,10,0.1) #[0, 0.1, 0.2, 0.3, ... ,9.9]
y = np.sin(x)

plt.plot(y, '--r', label='one')
plt.title('2 graphics')
plt.plot(2*y, 'g', lw=5, label='two', alpha=0.5)    # lw = linewidth
                                                    # alpha - прозрачность
# plot line y=0
plt.plot(0*np.ones(x.shape[0]), '--b')
plt.ylabel('y')
plt.xlabel('x')
# [xmin, xmax, ymin, ymax]
plt.axis([0, 100, -3, 3])

# ncol - n colomn in box
# bbox_to_anchor - % от левой и нижней границы
plt.legend(loc='lower left', facecolor='white', edgecolor='black', shadow=True, ncol=2,
           bbox_to_anchor=(0., 0.1))

plt.show()
```

```python
# ----------------------- Save Subplot without Display -------------------------------
import numpy as np
import matplotlib.pyplot as plt
plt.switch_backend('agg') # !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
x = np.arange(0,10,0.1) #[0, 0.1, 0.2, 0.3, ... ,9.9]
y = np.sin(x)
y2 = np.sin(2*x)

fig = plt.figure(figsize=(15, 10))   # this size of two graphics together
fig.suptitle('The title of the picture', fontsize=20, color='r')

plt.subplot(2, 1, 1)
plt.plot(y)
plt.title('3')

plt.subplot(2, 1, 2)
plt.plot(y2)
plt.title('32')

plt.savefig('out.png')

# comment this if you want to display result
plt.close(fig)

print('Saved succesfull')
```
